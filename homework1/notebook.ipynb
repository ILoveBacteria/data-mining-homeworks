## سوال ۳: بررسی و تبدیل نوع داده‌ها

**راهنما:** مقادیر گمشده در این مجموعه داده به صورت `NaN` ذخیره شده‌اند.

الف) بررسی نوع داده هر ستون

با استفاده از دستور `dtypes`، نوع داده هر ستون بررسی می‌شود تا وضعیت فعلی مشخص گردد.

ب) تعیین نوع داده بهینه برای هر ستون

برای اطمینان از صحت تحلیل‌ها و بهینه‌سازی مصرف حافظه، نوع داده هر ستون باید با دقت انتخاب شود. به موارد زیر باید توجه ویژه‌ای داشت:

-   **ستون‌های تاریخ (Date Columns):** ستون‌هایی که به صورت `object` ذخیره شده‌اند و حاوی اطلاعات زمانی هستند، باید به نوع داده تاریخ (مانند `datetime64[ns]`) تبدیل شوند تا عملیات مرتبط با تاریخ (مانند استخراج سال، ماه، روز، یا محاسبه بازه زمانی) به درستی قابل انجام باشند.
-   **ستون‌های دسته‌ای (Categorical Columns):** ستون‌هایی با کاردینالیتی پایین (تعداد مقادیر منحصر به فرد کم، مانند نام ایالت، روش پرداخت، وضعیت سفارش) باید به نوع داده `category` تبدیل شوند. این کار باعث کاهش چشمگیر مصرف حافظه می‌شود.
-   **ستون‌های عددی صحیح (Integer Numeric Columns):** ستون‌هایی که حاوی اعداد صحیح هستند اما به صورت `float` ذخیره شده‌اند (به دلیل وجود مقادیر `NaN` یا دقت اعشار)، در صورت امکان و پس از مدیریت مقادیر گمشده، باید به نوع داده صحیح (مانند `int64` یا انواع کوچکتر بسته به دامنه اعداد) تبدیل شوند تا حافظه کمتری اشغال کنند.

ج) انجام تبدیل نوع داده‌ها و گزارش حجم حافظه

تبدیل نوع داده‌ها انجام خواهد شد و حجم حافظه اشغال شده **قبل** و **بعد** از این تبدیل به طور دقیق گزارش می‌شود.

د) توضیح کاهش حافظه با تبدیل به نوع Category

تبدیل ستون‌های رشته‌ای با تکرار زیاد (مانند نام ایالت) به نوع `category` باعث کاهش حافظه می‌شود زیرا نحوه ذخیره‌سازی داخلی داده‌ها تغییر می‌کند.

-   **قبل از تبدیل (رشته‌ای):** هر مقدار رشته‌ای به طور مستقل و کامل در حافظه ذخیره می‌شود. اگر یک نام ایالت (مثلاً "تهران") بارها تکرار شده باشد، هر بار کل رشته "تهران" در حافظه نگهداری می‌شود که منجر به مصرف زیاد حافظه می‌گردد.
-   **بعد از تبدیل (Category):** در این حالت، ابتدا لیست تمام مقادیر منحصر به فرد (مثلاً "تهران"، "اصفهان"، "شیراز") یک بار ذخیره می‌شود. سپس هر ستون به جای ذخیره خود رشته‌ها، تنها یک شناسه (index) عددی را ذخیره می‌کند که به مقدار متناظرش در لیست مقادیر منحصر به فرد اشاره دارد. از آنجایی که این شناسه‌ها معمولاً اعداد صحیح کوچکی هستند، حجم حافظه مورد نیاز به شدت کاهش می‌یابد، به خصوص زمانی که تعداد تکرار مقادیر زیاد باشد.

</div>

## سوال ۵: پردازش داده‌های گمشده

الف) محاسبه و نمایش درصد مقادیر گمشده هر ستون

درصد مقادیر گمشده (`NaN`) برای هر ستون محاسبه شده و سپس با استفاده از یک نمودار مناسب (مانند نمودار میله‌ای) به صورت بصری نمایش داده می‌شود تا درک الگو و میزان پراکندگی داده‌های گمشده آسان‌تر گردد.

ب) دسته‌بندی ستون‌ها بر اساس درصد گمشدگی

ستون‌ها بر اساس درصد مقادیر گمشده به سه گروه تقسیم می‌شوند:

-   **گمشدگی کم (کمتر از ۵٪):** ستون‌هایی که درصد کمی از داده‌های آن‌ها گمشده است.
-   **گمشدگی متوسط (۵٪ تا ۴۰٪):** ستون‌هایی که درصد قابل توجهی اما نه بحرانی از داده‌های آن‌ها گمشده است.
-   **گمشدگی زیاد (بیش از ۴۰٪):** ستون‌هایی که بخش عمده‌ای از داده‌های آن‌ها گمشده است.

ج) پیشنهاد و پیاده‌سازی روش‌های مدیریت داده‌های گمشده

برای هر یک از سه گروه بالا، حداقل دو روش مختلف برای مدیریت داده‌های گمشده پیشنهاد خواهد شد. سپس، بر اساس معیارهایی مانند حفظ اطلاعات، کاهش سوگیری و حفظ حجم داده‌ها، یک روش برای هر گروه انتخاب شده، دلیل انتخاب آن ذکر و پیاده‌سازی می‌شود. روش‌های پیشنهادی می‌توانند شامل موارد زیر باشند:

*   **حذف ردیف (Row Deletion):** حذف کامل ردیف‌هایی که حاوی مقادیر گمشده هستند. این روش برای گروه "گمشدگی کم" مناسب است.
*   **حذف ستون (Column Deletion):** حذف کامل ستون‌هایی که درصد بالایی از داده‌هایشان گمشده است (برای گروه "گمشدگی زیاد").
*   **جایگزینی با میانه (Median Imputation):** جایگزینی مقادیر گمشده عددی با میانه ستون. این روش برای ستون‌های عددی در گروه "گمشدگی متوسط" مناسب است، زیرا کمتر تحت تاثیر داده‌های پرت قرار می‌گیرد.
*   **جایگزینی با مد (Mode Imputation):** جایگزینی مقادیر گمشده دسته‌ای (categorical) با مد (محتمل‌ترین مقدار) ستون. این روش نیز برای ستون‌های دسته‌ای در گروه "گمشدگی متوسط" یا "کم" کاربرد دارد.
*   **Fill-Forward / Backward Fill:** جایگزینی مقدار گمشده با مقدار معتبر قبلی (forward fill) یا بعدی (backward fill) در همان ستون. این روش برای داده‌های سری زمانی یا داده‌هایی که ترتیب ردیف‌ها اهمیت دارد، می‌تواند مفید باشد.

پس از انتخاب و پیاده‌سازی روش مناسب برای هر گروه، تغییرات اعمال شده و وضعیت داده‌های گمشده مجدداً بررسی خواهد شد.

د) بررسی الگوی گمشدگی داده‌ها

الگوهای خاص در گمشدگی داده‌ها بررسی می‌شوند. به عنوان مثال، مشخص می‌شود که آیا گمشدگی نظرات (reviews) با امتیاز خاصی از سفارشات مرتبط است (مثلاً آیا سفارشاتی با امتیاز پایین، نظر ندارند؟) یا خیر. این تحلیل به درک بهتر دلایل گمشدگی داده‌ها و تاثیر آن بر نتایج تحلیل کمک می‌کند.

</div>


```python
def display_percentage_of_nan(df):
    # Calculate NaN percentage for each column
    nan_counts = df.isnull().sum()
    percentages = (nan_counts / len(df)) * 100

    # Filter out columns with 0% NaN values
    plot_data = percentages[percentages > 0]

    # Create bar chart
    plot_data.plot(kind='bar')
    plt.title(f'Percentage of NaN values in columns')
    plt.xlabel('Columns')
    plt.ylabel('NaN Percentage (%)')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
```

## سوال 7 – شناسایی و تحلیل مقادیر پرت (Outliers (


الف( ستونهای عددی مهم )قیمت محصول، هزینه ارسال، مبلغ پرداخت، وزن محصول، زمان

دهید.
د( آیا تمام مقادیر پرت باید حذف شوند؟ برای هر ستون تصمیم مناسب بگیرید و دلیل آن را
ً خطا است یا ممکن است یک کاالی

توضیح دهید. )مثالً

```python
def plot_boxplot_with_outliers(df: pd.DataFrame, column_name: str):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
    axes[0].boxplot(df[column_name])
    axes[1].boxplot(df[column_name])
    axes[0].set_title(f"Box Plot of {column_name}")
    axes[1].set_title(f"Box Plot of {column_name}")
    axes[0].set_ylabel(column_name)
    axes[1].set_ylabel(column_name)

    # Calculate IQR and identify outliers
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    axes[0].set_ylim(Q1 - 2 * IQR, Q3 + 2 * IQR)
    num_outliers = len(outliers)
    print(f"Number of outliers in '{column_name}': {num_outliers} - {num_outliers * 100 / df.size}% of total data")
    plt.show()
```

```python

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers

    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance
```

```python
merged['distance_km'] = merged.apply(lambda row: haversine(row['geolocation_lat_x'], row['geolocation_lng_x'], row['geolocation_lat_y'], row['geolocation_lng_y']), axis=1)
```
