```python
import pandas as pd
```

```python
df = pd.read_csv('Groceries.csv')
df
```

```python
df.info()
```

```python
basket = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).reset_index()
basket['count'] = len(basket['itemDescription'])
basket
```

```python
dataset = basket['itemDescription'].tolist()
dataset
```

برای اجرای الگوریتم نیاز داریم تا تبدیل به one hot شود

```python
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
one_hot = encoder.fit(dataset).transform(dataset)
one_hot
```

ستون‌های موجود

```python
encoder.columns_
```

```python
df_one_hot = pd.DataFrame(one_hot, columns=encoder.columns_)
df_one_hot
```

```python
unique_items = len(df['itemDescription'].unique())
frequent = df.groupby('itemDescription').size().sort_values(ascending=False).to_frame('support').reset_index().head(10)
frequent["support_percent"] = frequent["support"] / len(basket) * 100
mean_item_in_basket = df.groupby(['Member_number', 'Date']).size().to_frame('size').reset_index()['size'].mean()


print(f'Count of customers = {len(basket['Member_number'].unique())}')
print(f'Unique items = {unique_items}')
print(f'Mean items in basket = {mean_item_in_basket}')
print('Top-10 most frequent items:')
print(frequent)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_06 = apriori(df_one_hot, min_support=0.06, use_colnames=True)
frequent_itemsets_06['support_count'] = frequent_itemsets_06['support'] * len(basket)
frequent_itemsets_06.sort_values('support', ascending=False)
```
