# Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: Better plot style
sns.set_theme(style="whitegrid")

# Columns to plot
columns = [
    "stress_level",
    "sleep_hours",
    "daily_social_media_hours",
    "age"
]

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

# Plot each column
for ax, col in zip(axes, columns):
    sns.histplot(
        data=df,
        x=col,
        bins=20,
        kde=True,
        color="steelblue",
        edgecolor="black",
        ax=ax
    )
    ax.set_title(f"Distribution of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Optional: Better plot style
sns.set_theme(style="whitegrid")

# Categorical columns
columns = [
    "platform_usage",
    "social_interaction_level"
]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot count bars
for ax, col in zip(axes, columns):
    sns.countplot(
        data=df,
        x=col,
        order=df[col].value_counts().index,  # Sort by frequency
        edgecolor="black",
        ax=ax
    )

    ax.set_title(f"Distribution of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

```python
# Count each class
counts = df["depression_label"].value_counts()

# Create pie chart

plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05] * len(counts),   # Slightly separate slices
    shadow=True
)

plt.title("Distribution of Depression Labels")
plt.axis("equal")  # Make the pie chart circular

plt.show()
```

# String to numerical value

```python
# Convert categorical columns to numerical

df["gender"] = df["gender"].map({
    "male": 0,
    "female": 1
})

df["platform_usage"] = df["platform_usage"].map({
    "Instagram": 0,
    "TikTok": 1,
    "Both": 2
})

df["social_interaction_level"] = df["social_interaction_level"].map({
    "low": 0,
    "medium": 1,
    "high": 2
})
```

```python
df
```

```python
# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Plot
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Heatmap", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
