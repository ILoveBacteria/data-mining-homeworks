# Normalization

```python
df
```

```python
y_column = 'depression_label'
X = df[list(set(df.columns) - {y_column})]
y = df[y_column]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X_normalize = scaler.transform(X)
X_normalize
```

# Imbalance data

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_normalize, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from imblearn.combine import SMOTETomek

smote_tomek = SMOTETomek(random_state=42)

X_train_balanced, y_train_balanced = smote_tomek.fit_resample(X_train, y_train)

print(y_train.value_counts())
print(y_train_balanced.value_counts())
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    precision_score,
    recall_score,
    f1_score
)
```

```python

model_log = LogisticRegression(random_state=42, max_iter=1000)
model_log.fit(X_train, y_train)

y_pred = model_log.predict(X_test)


print(f"Precision : {precision_score(y_test, y_pred):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred):.4f}")
print(f"F1-score  : {f1_score(y_test, y_pred):.4f}")

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python

model_log = LogisticRegression(random_state=42, max_iter=1000, class_weight="balanced")
model_log.fit(X_train, y_train)

y_pred = model_log.predict(X_test)


print(f"Precision : {precision_score(y_test, y_pred):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred):.4f}")
print(f"F1-score  : {f1_score(y_test, y_pred):.4f}")

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python

model_log = LogisticRegression(random_state=42, max_iter=1000)
model_log.fit(X_train_balanced, y_train_balanced)

y_pred = model_log.predict(X_test)


print(f"Precision : {precision_score(y_test, y_pred):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred):.4f}")
print(f"F1-score  : {f1_score(y_test, y_pred):.4f}")

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

results = []

best_k = None
best_f1 = -1
best_model_knn = None

for k in range(3, 12):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_balanced, y_train_balanced)

    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred)

    results.append({"k": k, "F1-score": f1})

    if f1 > best_f1:
        best_f1 = f1
        best_k = k
        best_model_knn = model

results_df = pd.DataFrame(results)

print(results_df)
print(f"\nBest k: {best_k}")
print(f"Best F1-score: {best_f1:.4f}")

```

```python

y_pred = best_model_knn.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=best_model_knn.classes_,
    yticklabels=best_model_knn.classes_
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title(f"Confusion Matrix (k={best_k})")
plt.tight_layout()
plt.show()

print(f"Precision : {precision_score(y_test, y_pred):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred):.4f}")
print(f"F1-score  : {f1_score(y_test, y_pred):.4f}")

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
train_f1 = []
test_f1 = []
depths = range(1, 10)

for depth in depths:
    model = DecisionTreeClassifier(
        max_depth=depth,
        random_state=42
    )

    model.fit(X_train_balanced, y_train_balanced)

    y_train_pred = model.predict(X_train_balanced)
    y_test_pred = model.predict(X_test)

    train_f1.append(f1_score(y_train_balanced, y_train_pred))
    test_f1.append(f1_score(y_test, y_test_pred))

plt.figure(figsize=(8, 5))

plt.plot(
    depths,
    train_f1,
    marker="o",
    linewidth=2,
    label="Train F1-score"
)

plt.plot(
    depths,
    test_f1,
    marker="s",
    linewidth=2,
    label="Test F1-score"
)

plt.xticks(depths)
plt.xlabel("max_depth")
plt.ylabel("F1-score")
plt.title("Decision Tree Performance vs. Tree Depth")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# SVM

By default does not compute the probability. I add `probability=True,` to it

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA

linear_svm = SVC(kernel="linear", probability=True, random_state=42)
rbf_svm = SVC(kernel="rbf", probability=True, random_state=42)

linear_svm.fit(X_train_balanced, y_train_balanced)
rbf_svm.fit(X_train_balanced, y_train_balanced)

y_pred_linear = linear_svm.predict(X_test)
y_pred_rbf = rbf_svm.predict(X_test)


```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(
    confusion_matrix(y_test, y_pred_linear),
    annot=True,
    fmt="d",
    cmap="Blues",
    ax=axes[0]
)
axes[0].set_title("Linear SVM")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

sns.heatmap(
    confusion_matrix(y_test, y_pred_rbf),
    annot=True,
    fmt="d",
    cmap="Blues",
    ax=axes[1]
)
axes[1].set_title("RBF SVM")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()


results = pd.DataFrame({
    "Model": ["Linear SVM", "RBF SVM"],
    "Precision": [
        precision_score(y_test, y_pred_linear),
        precision_score(y_test, y_pred_rbf)
    ],
    "Recall": [
        recall_score(y_test, y_pred_linear),
        recall_score(y_test, y_pred_rbf)
    ],
    "F1-score": [
        f1_score(y_test, y_pred_linear),
        f1_score(y_test, y_pred_rbf)
    ]
})

print(results)

print("\nLinear SVM Classification Report\n")
print(classification_report(y_test, y_pred_linear))

print("\nRBF SVM Classification Report\n")
print(classification_report(y_test, y_pred_rbf))
```

PCA requires variables be in Standardize form. So rescale all features...

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

scaler_std = StandardScaler()

X_standard = scaler_std.fit_transform(X)

X_train_std, X_test_std, y_train_std, y_test_std = train_test_split(
    X_standard,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

smote_tomek_std = SMOTETomek(random_state=42)

X_train_std_balanced, y_train_std_balanced = smote_tomek_std.fit_resample(
    X_train_std,
    y_train_std
)

pca = PCA(n_components=2)

X_train_std_pca = pca.fit_transform(X_train_std_balanced)
X_test_std_pca = pca.transform(X_test_std)

linear_svm_pca = SVC(kernel="linear", random_state=42)
rbf_svm_pca = SVC(kernel="rbf", random_state=42)

linear_svm_pca.fit(X_train_std_pca, y_train_std_balanced)
rbf_svm_pca.fit(X_train_std_pca, y_train_std_balanced)

x_min = X_train_std_pca[:, 0].min() - 1
x_max = X_train_std_pca[:, 0].max() + 1
y_min = X_train_std_pca[:, 1].min() - 1
y_max = X_train_std_pca[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, model, title in zip(
    axes,
    [linear_svm_pca, rbf_svm_pca],
    ["Linear SVM", "RBF SVM"]
):
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")

    ax.scatter(
        X_test_std_pca[:, 0],
        X_test_std_pca[:, 1],
        c=y_test_std,
        cmap="coolwarm",
        edgecolors="k",
        s=40
    )

    ax.set_title(title)
    ax.set_xlabel("Principal Component 1")
    ax.set_ylabel("Principal Component 2")

plt.tight_layout()
plt.show()
```

Also check the decision boundaries of LogisticRegression

```python
pca = PCA(n_components=2)

X_train_std_pca = pca.fit_transform(X_train_std_balanced)
X_test_std_pca = pca.transform(X_test_std)

logistic_pca = LogisticRegression(
    random_state=42,
    max_iter=1000
)

logistic_pca.fit(X_train_std_pca, y_train_std_balanced)

x_min = X_train_std_pca[:, 0].min() - 1
x_max = X_train_std_pca[:, 0].max() + 1
y_min = X_train_std_pca[:, 1].min() - 1
y_max = X_train_std_pca[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

Z = logistic_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))

plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")

plt.scatter(
    X_test_std_pca[:, 0],
    X_test_std_pca[:, 1],
    c=y_test_std,
    cmap="coolwarm",
    edgecolors="k",
    s=40
)

plt.title("Logistic Regression Decision Boundary (PCA)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.tight_layout()
plt.show()
```

# XGBoost

```python
from xgboost import XGBClassifier, plot_tree
from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.1, 0.2],
    "n_estimators": [50, 100, 200]
}

xgb = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    random_state=42
)

grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring="f1_macro",
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train_balanced, y_train_balanced)

print("Best Parameters:")
print(grid.best_params_)

print(f"\nBest CV F1_macro: {grid.best_score_:.4f}")

best_model_xgb = grid.best_estimator_

y_pred = best_model_xgb.predict(X_test)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

print(f"Precision : {precision_score(y_test, y_pred):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred):.4f}")
print(f"F1-score  : {f1_score(y_test, y_pred):.4f}")

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)
plt.title("XGBoost Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

# Model evaluation and selection

```python
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score

models = {
    "Logistic Regression": model_log,
    "KNN": best_model_knn,
    "Decision Tree": model_tree,
    "Linear SVM": linear_svm,
    "RBF SVM": rbf_svm,
    "Random Forest": rf_model,
    "XGBoost": best_model_xgb
}

results = []

plt.figure(figsize=(8, 6))

for name, model in models.items():

    y_pred = model.predict(X_test)
    y_score = model.predict_proba(X_test)[:, 1]

    fpr, tpr, _ = roc_curve(y_test, y_score)
    auc = roc_auc_score(y_test, y_score)

    plt.plot(fpr, tpr, linewidth=2, label=f"{name} (AUC={auc:.3f})")

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-score": f1_score(y_test, y_pred),
        "AUC": auc
    })

plt.plot([0, 1], [0, 1], "k--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

results_df = pd.DataFrame(results)

print(results_df.sort_values("F1-score", ascending=False))
```
